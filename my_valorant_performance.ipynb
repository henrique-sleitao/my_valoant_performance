```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
from PIL import Image
import math
import os
```

```python
df_my_statistics = pd.read_csv("/content/my_statistics_iron3.csv")
df_my_statistics.head()
```

```python
df_global_statistics = pd.read_csv("/content/global_statistics_iron3.csv")
df_global_statistics.head()
```

This project has the objective to show some metrics of my gameplay from Valorant (It's a really popular fps game on the world) and compare it with the rest of the world on my rank witch is Iron 3.

#My kills average by agent per match


```python
background_image = Image.open("/content/pattern3.jpg")

df_sorted = df_my_statistics.sort_values(by='kills', ascending=False)
custom_palette = sns.color_palette("Reds_r", len(df_sorted))

x_min = -0.5
x_max = len(df_sorted) - 0.5
y_max = int(df_sorted['kills'].max()) + 5

plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.set_facecolor('none')

plt.imshow(background_image, aspect='auto', extent=(x_min, x_max, 0, y_max), alpha=0.8)

ax = sns.barplot(x='agents', y='kills', data=df_sorted, palette=custom_palette, alpha=0.9)
plt.title('My kills average by agent per match')
plt.xlabel('agents')
plt.ylabel('kills')

plt.yticks(range(0, y_max + 1, 5))

for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f'{y:.0f}', (x, y), ha='center', va='bottom', color='white', fontsize=12, fontweight='bold')

plt.grid(visible=False)

plt.xlim(x_min, x_max)
plt.ylim(0, y_max)


brimstone = Image.open('/content/sample_data/imagens/brimstone.png')

fig_width = 2
fig_height = 5

plt.figure(figsize=(fig_width, fig_height))

plt.imshow(brimstone)
plt.axis('off')

plt.show()

```

#My kills/deaths ratio by agent

```python
background_image = Image.open("/content/pattern3.jpg")

df_sorted = df_my_statistics.sort_values(by='kills/deaths', ascending=False)
custom_palette = sns.color_palette("Reds_r", len(df_sorted))

x_min = -0.5
x_max = len(df_sorted) - 0.5
y_max = int(df_sorted['kills/deaths'].max()) + 1

plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.set_facecolor('none')

plt.imshow(background_image, aspect='auto', extent=(x_min, x_max, 0, 1.5), alpha=0.8)
ax = sns.barplot(x='agents', y='kills/deaths', data=df_sorted, palette=custom_palette)
plt.title('My kills/deaths ratio by agent')
plt.xlabel('agents')
plt.ylabel('kills/deaths ratio')


plt.yticks(np.arange(0, 1.75, 0.25))

for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f'{y:.2f}', (x, y), ha='center', va='bottom', color='white', fontsize=12, fontweight='bold')


plt.grid(visible=False)

plt.xlim(x_min, x_max)
plt.ylim(0, 1.5)


brimstone = Image.open('/content/sample_data/imagens/brimstone.png')

fig_width = 2
fig_height = 5

plt.figure(figsize=(fig_width, fig_height))

plt.imshow(brimstone)
plt.axis('off')

plt.show()
```

On kills/deaths ratio we can see a good performance when this metric is above 1 as this shows us that during the match the character killed more than died. So it can be concluded that my best performance as a player was using the character Brimstone and then in second place to Raze, but this performance does not mean that it was decisive for the player to emerge victorious from the match. So let's see on the next chart witch character could be more useful to make me win more matches.

#My win% by agent

```python
background_image = Image.open("/content/pattern3.jpg")

df_sorted = df_my_statistics.sort_values(by='win%', ascending=False)
custom_palette = sns.color_palette("Reds_r", len(df_sorted))

x_min = -0.5
x_max = len(df_sorted) - 0.5
y_max = int(df_sorted['win%'].max()) + 1

plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.set_facecolor('none')

plt.imshow(background_image, aspect='auto', extent=(x_min, x_max, 0, 100), alpha=0.8)
ax = sns.barplot(x='agents', y='win%', data=df_sorted, palette=custom_palette)
plt.title('My win% by agent')
plt.xlabel('agents')
plt.ylabel('win%')

plt.yticks(range(0, 110, 25))
plt.ylim(0, 100)

for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f'{y:.2f}', (x, y), ha='center', va='bottom', color='white', fontsize=12, fontweight='bold')

plt.grid(visible=False)

plt.xlim(x_min, x_max)
plt.ylim(0, 100)


raze = Image.open('/content/sample_data/imagens/raze.png')

fig_width = 2
fig_height = 5

plt.figure(figsize=(fig_width, fig_height))

plt.imshow(raze)
plt.axis('off')

plt.show()
```

So there it is, here we can see that when using the character Raze I tend to be victorious in matches, while with the rest of the characters I chose I tend to lose more than I win. But she is not my favorite character.

#My KDA

```python
pivot_table = df_my_statistics.pivot_table(index='agents', values=['kills', 'deaths', 'assists'])

pivot_table = pivot_table.transpose()

pivot_table = pivot_table.sort_index(ascending=False)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='RdYlBu_r', annot=True, fmt=".1f", linewidths=0.5)

plt.title('My KDA')
plt.xlabel('Estatístics')
plt.ylabel('Agents')

plt.show()

```

In this hitmap graph we can compare my average performance in kills/deaths/assists (KDA) according to the agents that I already chose in the game. Where I usually take advantage in Kills and Assists with the character Brimstone but in Deaths my performance is better using the character Raze.

#Global Iron 3 KDA

```python
pivot_table = df_global_statistics.pivot_table(index='agents', values=['kills', 'deaths', 'assists'])

pivot_table = pivot_table.transpose()

pivot_table = pivot_table.sort_index(ascending=False)

plt.figure(figsize=(19, 6))
sns.heatmap(pivot_table, cmap='RdYlBu_r', annot=True, fmt=".1f", linewidths=0.5)

plt.title('Global Iron 3 KDA')
plt.xlabel('Estatístics')
plt.ylabel('Agents')

plt.show()
```

When we look at the global performance things are a little bit diferent. We can see 3 diferent leaders acording to each indicator the best killer character is Reyna, the best assistant is KAY/O and the mosta hard to kill is Killjoy.

#My kills/deaths ratio vs wining ratio

```python

agent_means = df_my_statistics.groupby('agents').mean()
agent_means['kills/deaths'] = agent_means['kills'] / agent_means['deaths'] * 100

plt.figure(figsize=(10, 6))
ax = sns.heatmap(agent_means[['kills/deaths', 'win%']], cmap='RdYlBu', annot=True, fmt=".1f", linewidths=0.5)

plt.title('Média de Kills/Deaths vs. Média de Win% por Agente')
plt.xlabel('Estatísticas')
plt.ylabel('Agente')

cbar = ax.collections[0].colorbar
cbar.set_ticks([0, 100])
cbar.set_ticklabels(['0%', '100%'])

plt.show()
```

Here we can see that good individual performances do not necessarily imply an increase in the number of wins on Iron 3 level. In order to obtain more victories, it is necessary that all team members fulfill their duties during the match.
